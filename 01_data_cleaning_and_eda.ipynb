### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Plot style
sns.set(style="whitegrid")
```

### Drop or Impute Null Values

```python
# Remove rows with missing income
df = df[df['Income'].notna()]
```

```python
df.isnull().sum()
```

```python
df['Marital_Status'] = df['Marital_Status'].replace({'YOLO': 'Other', 'Absurd': 'Other'})
```

```python
df['Marital_Status'].unique()
```

### Feature Engineering

```python
# Combine campaign response flags
campaign_cols = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']
df['TotalAcceptedCmp'] = df[campaign_cols].sum(axis=1)

# Age
df['Age'] = 2025 - df['Year_Birth']

# Categorize age
df['AgeGroup'] = pd.cut(df['Age'], bins=[18, 30, 45, 60, 100], labels=['18–30', '31–45', '46–60', '60+'])

# Total Spend
spend_cols = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df['TotalSpend'] = df[spend_cols].sum(axis=1)
```

```python
df = df[df['Age'].notnull()]
```

### Basic EDA (Seaborn + Matplotlib)

```python
# Distribution of Age
plt.figure(figsize=(10, 5))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Customer Age Distribution')

# Campaign responses by Age Group
plt.figure(figsize=(10, 6))
sns.barplot(x='AgeGroup', y='TotalAcceptedCmp', data=df, estimator=np.mean)
plt.title('Avg Campaign Acceptance by Age Group')

# Response by Marital Status
plt.figure(figsize=(10, 6))
sns.boxplot(x='Marital_Status', y='TotalAcceptedCmp', data=df)
plt.title('Campaign Response by Marital Status')
```
